### <font color='264CC7'> Análisis estadístico básico </font>

     <ul>

  <li>Variables numéricas:</li>
    <ul>
      <li>Calcula estadísticos descriptivos básicos como media, mediana, desviación estándar, entre otros.</li>

Sugerencia: Puedes usar la librería 'pandas' y sus funciones 'describe' y 'value_counts', así como las funciones 'bar' e 'hist' de matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# 1. Cargar el dataset
# ==============================
url = "https://raw.githubusercontent.com/rupakc/UCI-Data-Analysis/refs/heads/master/Boston%20Housing%20Dataset/Boston%20Housing/housing.data"
column_names = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
    "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"
]
df = pd.read_csv(url, delim_whitespace=True, names=column_names)

# ==============================
# 2. Variables categóricas
# ==============================
# Frecuencia de CHAS y RAD
print("Frecuencia de CHAS:")
print(df["CHAS"].value_counts())
print("\nFrecuencia de RAD:")
print(df["RAD"].value_counts())

# Gráfico de barras para CHAS
plt.figure(figsize=(6,4))
df["CHAS"].value_counts().plot(kind="bar", color="skyblue")
plt.title("Frecuencia de CHAS")
plt.xlabel("CHAS")
plt.ylabel("Frecuencia")
plt.show()

# Gráfico de barras para RAD
plt.figure(figsize=(8,4))
df["RAD"].value_counts().plot(kind="bar", color="orange")
plt.title("Frecuencia de RAD")
plt.xlabel("RAD")
plt.ylabel("Frecuencia")
plt.show()

# ==============================
# 3. Variables numéricas
# ==============================
# Estadísticos descriptivos básicos
print("\nEstadísticos descriptivos:")
print(df.describe())

# Mediana y desviación estándar
print("\nMediana de cada variable:")
print(df.median())
print("\nDesviación estándar de cada variable:")
print(df.std())

# ==============================
# 4. Histogramas
# ==============================
variables = ["INDUS", "AGE", "MEDV"]
for var in variables:
    plt.figure(figsize=(6,4))
    df[var].hist(bins=20, color="green", edgecolor="black")
    plt.title(f"Histograma de {var}")
    plt.xlabel(var)
    plt.ylabel("Frecuencia")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# 1. Cargar el dataset
# ==============================
url = "https://raw.githubusercontent.com/rupakc/UCI-Data-Analysis/refs/heads/master/Boston%20Housing%20Dataset/Boston%20Housing/housing.data"
column_names = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
    "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"
]
df = pd.read_csv(url, delim_whitespace=True, names=column_names)

# ==============================
# 2. Histograma de MEDV separado por CHAS
# ==============================
plt.figure(figsize=(8,6))

# Histograma para CHAS = 0
plt.hist(df[df["CHAS"] == 0]["MEDV"], bins=20, alpha=0.6, density=True, color="blue", label="CHAS = 0")

# Histograma para CHAS = 1
plt.hist(df[df["CHAS"] == 1]["MEDV"], bins=20, alpha=0.6, density=True, color="red", label="CHAS = 1")

# Configuración del gráfico
plt.title("Distribución de MEDV según CHAS")
plt.xlabel("MEDV (Valor medio de las casas en $1000s)")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ==============================
# 1. Cargar el dataset
# ==============================
url = "https://raw.githubusercontent.com/rupakc/UCI-Data-Analysis/refs/heads/master/Boston%20Housing%20Dataset/Boston%20Housing/housing.data"
column_names = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
    "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"
]
df = pd.read_csv(url, delim_whitespace=True, names=column_names)

# ==============================
# 2. Calcular correlaciones con MEDV
# ==============================
correlaciones = df.corr()["MEDV"].sort_values(ascending=False)
print("Correlación de cada atributo con MEDV:")
print(correlaciones)

# ==============================
# 3. Visualizar con mapa de calor
# ==============================
plt.figure(figsize=(8,6))
sns.heatmap(df.corr()[["MEDV"]].sort_values(by="MEDV", ascending=False),
            annot=True, cmap="coolwarm")
plt.title("Correlación de atributos con MEDV")
plt.show()
```
